```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the describe function
data.describe()
```

```python
# We have noticed that the column itemprice contains 3 empty slots.
# We need to fill these empty slots with a value and in this case, we will use the mean of the column to fill the empty slots.
data["itemprice"] = data["itemprice"].fillna(55)
# Describe the data agian
data.describe()
```

```python
# Get the number of enteries in a given column
data.groupby("gender").size()
```

```python
# Combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of males to females
sns.countplot(x = "product", data=data, palette="dark")
plt.title("Distribution of Products")
plt.xticks(rotation= 90)
plt.ylabel("Total Count Of Products")
```

```python
# Below is a count plot that has 2 columns
sns.countplot(x= "gender", hue="city", palette="Reds", data=data)
```

```python
# Below is a pie plot
data["payment"].value_counts().plot.pie(autopct='%1.2f%%')
```
